## Plotting The Bar Chart for Male V/s Female for Survived and Not Survived

```python
survival_by_sex = df.groupby(["Sex", "Survived"]).size().unstack(fill_value=0)
survival_by_sex.plot(kind='bar', stacked=False, color=['red', 'green'])

plt.title("Survival Counts by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.legend(["Not Survived", "Survived"])
plt.xticks(rotation=0)

# Add counts on top of each bar for clarity
for idx, row in enumerate(survival_by_sex.values):
    for i, count in enumerate(row):
        plt.text(idx - 0.15 + i*0.3, count + 5, str(count), ha='center', va='bottom', fontsize=9)
plt.show()
```

## Plotting the scatter flot for comparing the survival rates using the fare price

```python
plt.figure(figsize=(8, 6))
plt.scatter(train['Survived'], train['Fare'], color='blue')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Fare')
plt.title('Scatter Plot of Survival vs Fare (Titanic Dataset)')
plt.xticks([0 , 1 ])
plt.show()
# It shows that the passenge who has paid more has a greater chance of survival.
```

## Plotting the bar plot for the survival rate for different passenger class

```python
classwise_survival = train.groupby('Pclass')['Survived'].mean()

classwise_survival.plot(kind='bar', color=['gold', 'silver', 'brown'])
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.xticks([0,1,2], ['1st', '2nd', '3rd'])
plt.ylim(0,1)
plt.show()
```
